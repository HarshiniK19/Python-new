```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
da = pd.read_csv("Sales_Data.csv")
```

```python
# from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
da['Particulars']= le.fit_transform(da['Particulars'])
da['GSTIN/UIN']= le.fit_transform(da['GSTIN/UIN'])
da['Narration']= le.fit_transform(da['Narration'])
da
```

```python
da = da.drop(['GSTIN/UIN','Voucher Ref.','Alt. Units','Voucher No.'],axis=1)
da
```

```python
da.dropna(inplace=True)
```

```python
da.isnull().sum()
```

```python
x=da.drop(['Gross Total','Date','Voucher No.','Narration','GSTIN/UIN',],axis=1)
y=da['Gross Total']
da
```

```python
y
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
x = SS.fit_transform(x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
lr=LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
print(pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,pred)
print(r2)

from sklearn.metrics import mean_absolute_error
m=mean_absolute_error(y_test,pred)
print(m)
```

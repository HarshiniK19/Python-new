```python
# from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
da['Weather']= le.fit_transform(da['Weather'])
da['Vehicle_Type']= le.fit_transform(da['Vehicle_Type'])
da['Traffic_Level']= le.fit_transform(da['Traffic_Level'])
da['Time_of_Day']= le.fit_transform(da['Time_of_Day'])
da
```

```python
da.drop(['Courier_Experience_yrs'],axis=1)
da
```

```python
da.isnull().sum()
```

```python
da.dropna(inplace=True)
```

```python
x=da.drop(['Delivery_Time_min','Order_ID','Courier_Experience_yrs'],axis=1)
y=da['Delivery_Time_min']
da
```

```python
y
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
lr=LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
print(pred)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,pred)
print(r2)

from sklearn.metrics import mean_absolute_error
m=mean_absolute_error(y_test,pred)
print(m)
```
